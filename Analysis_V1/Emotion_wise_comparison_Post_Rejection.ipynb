# Emotion Comprison

## Overall

```python
from scipy.stats import kruskal
# Define the emotion labels for each set
emotion_labels = {
    'Set A': ['Sad', 'Happy', 'Angry', 'Neutral', 'Happy', 'Sad', 'Neutral', 'Sad', 'Happy', 'Sad',
              'Angry', 'Sad', 'Happy', 'Neutral', 'Angry', 'Happy', 'Angry', 'Neutral', 'Angry', 'Neutral',
              'Angry', 'Neutral', 'Happy', 'Neutral', 'Sad'],
    'Set B': ['Sad', 'Neutral', 'Angry', 'Neutral', 'Angry', 'Happy', 'Angry', 'Happy', 'Sad', 'Happy',
              'Angry', 'Happy', 'Sad', 'Happy', 'Sad', 'Neutral', 'Sad', 'Angry', 'Neutral', 'Happy',
              'Neutral', 'Sad', 'Angry', 'Sad', 'Neutral'],
    'Set C': ['Neutral', 'Happy', 'Sad', 'Happy', 'Angry', 'Happy', 'Angry', 'Sad', 'Happy', 'Neutral',
              'Happy', 'Sad', 'Sad', 'Neutral', 'Angry', 'Sad', 'Neutral', 'Angry', 'Happy', 'Neutral',
              'Sad', 'Happy', 'Angry', 'Neutral', 'Angry'],
    'Set D': ['Angry', 'Sad', 'Happy', 'Sad', 'Neutral', 'Happy', 'Angry', 'Sad', 'Angry', 'Neutral',
              'Sad', 'Neutral', 'Sad', 'Happy', 'Angry', 'Happy', 'Sad', 'Neutral', 'Angry', 'Happy',
              'Neutral', 'Angry', 'Happy', 'Neutral', 'Angry']
}

# Function to extract valence and arousal ratings based on emotion labels and set name
def extract_ratings_by_set(df, labels, set_name):
    valence_ratings = {'Happy': [], 'Sad': [], 'Angry': [], 'Neutral': []}
    arousal_ratings = {'Happy': [], 'Sad': [], 'Angry': [], 'Neutral': []}

    for i, emotion in enumerate(labels):
        val_col = f'q{i+1}_val_{set_name.lower()}'
        arous_col = f'q{i+1}_arous_{set_name.lower()}'

        valence_ratings[emotion].extend(df[val_col])
        arousal_ratings[emotion].extend(df[arous_col])

    return valence_ratings, arousal_ratings

# Extract ratings for each set
valence_ratings_A, arousal_ratings_A = extract_ratings_by_set(df1, emotion_labels['Set A'], 'SetA')
valence_ratings_B, arousal_ratings_B = extract_ratings_by_set(df2, emotion_labels['Set B'], 'SetB')
valence_ratings_C, arousal_ratings_C = extract_ratings_by_set(df3, emotion_labels['Set C'], 'SetC')
valence_ratings_D, arousal_ratings_D = extract_ratings_by_set(df4, emotion_labels['Set D'], 'SetD')

# Combine ratings from all sets
combined_valence_ratings = {'Happy': valence_ratings_A['Happy'] + valence_ratings_B['Happy'] + valence_ratings_C['Happy'] + valence_ratings_D['Happy'],
                            'Sad': valence_ratings_A['Sad'] + valence_ratings_B['Sad'] + valence_ratings_C['Sad'] + valence_ratings_D['Sad'],
                            'Angry': valence_ratings_A['Angry'] + valence_ratings_B['Angry'] + valence_ratings_C['Angry'] + valence_ratings_D['Angry'],
                            'Neutral': valence_ratings_A['Neutral'] + valence_ratings_B['Neutral'] + valence_ratings_C['Neutral'] + valence_ratings_D['Neutral']}

combined_arousal_ratings = {'Happy': arousal_ratings_A['Happy'] + arousal_ratings_B['Happy'] + arousal_ratings_C['Happy'] + arousal_ratings_D['Happy'],
                            'Sad': arousal_ratings_A['Sad'] + arousal_ratings_B['Sad'] + arousal_ratings_C['Sad'] + arousal_ratings_D['Sad'],
                            'Angry': arousal_ratings_A['Angry'] + arousal_ratings_B['Angry'] + arousal_ratings_C['Angry'] + arousal_ratings_D['Angry'],
                            'Neutral': arousal_ratings_A['Neutral'] + arousal_ratings_B['Neutral'] + arousal_ratings_C['Neutral'] + arousal_ratings_D['Neutral']}

# Redefine the Kruskal-Wallis test function
def perform_kruskal_wallis(ratings, emotion_pairs):
    results = {}
    for pair in emotion_pairs:
        emotion1, emotion2 = pair
        stat, p_value = kruskal(ratings[emotion1], ratings[emotion2])
        results[pair] = (stat, p_value)
    return results

# Redefine emotion pairs for comparison
emotion_pairs = [('Happy', 'Sad'), ('Happy', 'Angry'), ('Happy', 'Neutral'),
                 ('Sad', 'Angry'), ('Sad', 'Neutral'), ('Angry', 'Neutral')]

# Perform Kruskal-Wallis tests
valence_results = perform_kruskal_wallis(combined_valence_ratings, emotion_pairs)
arousal_results = perform_kruskal_wallis(combined_arousal_ratings, emotion_pairs)

# Display the results
dataframe=pd.DataFrame({'Valence': valence_results, 'Arousal': arousal_results})
# tools.display_dataframe_to_user(name="Valence and Arousal Kruskal-Wallis Test Results",
valence_results, arousal_results
```

## Setwise Emotion Comparison

```python
# Perform Kruskal-Wallis tests for each set individually

# Define a function to perform tests for a single set
def perform_tests_for_set(valence_ratings, arousal_ratings, emotion_pairs):
    valence_results = perform_kruskal_wallis(valence_ratings, emotion_pairs)
    arousal_results = perform_kruskal_wallis(arousal_ratings, emotion_pairs)
    return valence_results, arousal_results

# Perform tests for each set
valence_results_A, arousal_results_A = perform_tests_for_set(valence_ratings_A, arousal_ratings_A, emotion_pairs)
valence_results_B, arousal_results_B = perform_tests_for_set(valence_ratings_B, arousal_ratings_B, emotion_pairs)
valence_results_C, arousal_results_C = perform_tests_for_set(valence_ratings_C, arousal_ratings_C, emotion_pairs)
valence_results_D, arousal_results_D = perform_tests_for_set(valence_ratings_D, arousal_ratings_D, emotion_pairs)

# Display the results for each set
# Set A valence and arousal
dataframe=pd.DataFrame({'Valence': valence_results_A, 'Arousal': arousal_results_A})
# Set B Valence and arousal
dataframe=pd.DataFrame({'Valence': valence_results_B, 'Arousal': arousal_results_B})

# Set C Valence and arousal
dataframe=pd.DataFrame({'Valence': valence_results_C, 'Arousal': arousal_results_C})

# Set D Valence and arousal
dataframe=pd.DataFrame({'Valence': valence_results_D, 'Arousal': arousal_results_D})

valence_results_A, arousal_results_A, valence_results_B, arousal_results_B, valence_results_C, arousal_results_C, valence_results_D, arousal_results_D
```

## Subtractive Comparison

### overall

```python
# Function to calculate differences from neutral ratings
def calculate_differences(ratings):
    differences = {}
    for emotion in ['Happy', 'Sad', 'Angry']:
        differences[emotion] = [rating - neutral for rating, neutral in zip(ratings[emotion], ratings['Neutral'])]
    return differences

# Calculate differences from neutral ratings for valence and arousal
valence_differences = calculate_differences(combined_valence_ratings)
arousal_differences = calculate_differences(combined_arousal_ratings)

# Define the emotion pairs for subtractive comparison
subtractive_pairs = [('Happy', 'Sad'), ('Happy', 'Angry'), ('Sad', 'Angry')]

# Perform Kruskal-Wallis tests on the differences
valence_subtractive_results = perform_kruskal_wallis(valence_differences, subtractive_pairs)
arousal_subtractive_results = perform_kruskal_wallis(arousal_differences, subtractive_pairs)

# Create a DataFrame for the subtractive comparison results
valence_subtractive_df = pd.DataFrame.from_dict(valence_subtractive_results, orient='index', columns=['Test Statistic', 'p-value'])
arousal_subtractive_df = pd.DataFrame.from_dict(arousal_subtractive_results, orient='index', columns=['Test Statistic', 'p-value'])

# Combine the results into a single DataFrame
subtractive_results_df = pd.DataFrame({
    'Valence Test Statistic': valence_subtractive_df['Test Statistic'],
    'Valence p-value': valence_subtractive_df['p-value'],
    'Arousal Test Statistic': arousal_subtractive_df['Test Statistic'],
    'Arousal p-value': arousal_subtractive_df['p-value']
})

subtractive_results_df.reset_index(inplace=True)
subtractive_results_df.rename(columns={'index': 'Emotion Pair'}, inplace=True)

# Display the DataFrame to the user
dataframe=subtractive_results_df

subtractive_results_df
```

### Setwise

```python

# Function to calculate differences from neutral ratings for a specific set correctly
def calculate_set_differences_correct(df, labels, set_name):
    valence_differences = {'Happy': [], 'Sad': [], 'Angry': []}
    arousal_differences = {'Happy': [], 'Sad': [], 'Angry': []}

    for i, emotion in enumerate(labels):
        if emotion != 'Neutral':
            val_col = f'q{i+1}_val_{set_name.lower()}'
            arous_col = f'q{i+1}_arous_{set_name.lower()}'

            # Find the corresponding neutral rating for this position
            neutral_position = labels.index('Neutral')
            neutral_val_col = f'q{neutral_position+1}_val_{set_name.lower()}'
            neutral_arous_col = f'q{neutral_position+1}_arous_{set_name.lower()}'

            if emotion in valence_differences:
                diff_val = df[val_col] - df[neutral_val_col]
                diff_arous = df[arous_col] - df[neutral_arous_col]

                valence_differences[emotion].extend(diff_val)
                arousal_differences[emotion].extend(diff_arous)

    return valence_differences, arousal_differences

# Calculate differences for each set individually
valence_differences_A_correct, arousal_differences_A_correct = calculate_set_differences_correct(df1, emotion_labels['Set A'], 'SetA')
valence_differences_B_correct, arousal_differences_B_correct = calculate_set_differences_correct(df2, emotion_labels['Set B'], 'SetB')
valence_differences_C_correct, arousal_differences_C_correct = calculate_set_differences_correct(df3, emotion_labels['Set C'], 'SetC')
valence_differences_D_correct, arousal_differences_D_correct = calculate_set_differences_correct(df4, emotion_labels['Set D'], 'SetD')

# Check for identical values and handle them
def perform_safe_kruskal_wallis(differences, pairs):
    results = {}
    for pair in pairs:
        emotion1, emotion2 = pair
        if len(set(differences[emotion1])) == 1 and len(set(differences[emotion2])) == 1:
            results[pair] = (None, None)  # Identical values, so no test result
        else:
            stat, p_value = kruskal(differences[emotion1], differences[emotion2])
            results[pair] = (stat, p_value)
    return results


# Perform Kruskal-Wallis tests on the differences for each set
valence_results_A_diff_correct = perform_safe_kruskal_wallis(valence_differences_A_correct, subtractive_pairs)
arousal_results_A_diff_correct = perform_safe_kruskal_wallis(arousal_differences_A_correct, subtractive_pairs)
valence_results_B_diff_correct = perform_safe_kruskal_wallis(valence_differences_B_correct, subtractive_pairs)
arousal_results_B_diff_correct = perform_safe_kruskal_wallis(arousal_differences_B_correct, subtractive_pairs)
valence_results_C_diff_correct = perform_safe_kruskal_wallis(valence_differences_C_correct, subtractive_pairs)
arousal_results_C_diff_correct = perform_safe_kruskal_wallis(arousal_differences_C_correct, subtractive_pairs)
valence_results_D_diff_correct = perform_safe_kruskal_wallis(valence_differences_D_correct, subtractive_pairs)
arousal_results_D_diff_correct = perform_safe_kruskal_wallis(arousal_differences_D_correct, subtractive_pairs)

# Create DataFrames for each set's subtractive comparison results
valence_diff_df_A_correct = pd.DataFrame.from_dict(valence_results_A_diff_correct, orient='index', columns=['Test Statistic', 'p-value'])
arousal_diff_df_A_correct = pd.DataFrame.from_dict(arousal_results_A_diff_correct, orient='index', columns=['Test Statistic', 'p-value'])

valence_diff_df_B_correct = pd.DataFrame.from_dict(valence_results_B_diff_correct, orient='index', columns=['Test Statistic', 'p-value'])
arousal_diff_df_B_correct = pd.DataFrame.from_dict(arousal_results_B_diff_correct, orient='index', columns=['Test Statistic', 'p-value'])

valence_diff_df_C_correct = pd.DataFrame.from_dict(valence_results_C_diff_correct, orient='index', columns=['Test Statistic', 'p-value'])
arousal_diff_df_C_correct = pd.DataFrame.from_dict(arousal_results_C_diff_correct, orient='index', columns=['Test Statistic', 'p-value'])

valence_diff_df_D_correct = pd.DataFrame.from_dict(valence_results_D_diff_correct, orient='index', columns=['Test Statistic', 'p-value'])
arousal_diff_df_D_correct = pd.DataFrame.from_dict(arousal_results_D_diff_correct, orient='index', columns=['Test Statistic', 'p-value'])

# Combine the results into a single DataFrame for each set
results_df_A_correct = pd.DataFrame({
    'Valence Test Statistic': valence_diff_df_A_correct['Test Statistic'],
    'Valence p-value': valence_diff_df_A_correct['p-value'],
    'Arousal Test Statistic': arousal_diff_df_A_correct['Test Statistic'],
    'Arousal p-value': arousal_diff_df_A_correct['p-value']
})
results_df_A_correct.reset_index(inplace=True)
results_df_A_correct.rename(columns={'index': 'Emotion Pair'}, inplace=True)

results_df_B_correct = pd.DataFrame({
    'Valence Test Statistic': valence_diff_df_B_correct['Test Statistic'],
    'Valence p-value': valence_diff_df_B_correct['p-value'],
    'Arousal Test Statistic': arousal_diff_df_B_correct['Test Statistic'],
    'Arousal p-value': arousal_diff_df_B_correct['p-value']
})
results_df_B_correct.reset_index(inplace=True)
results_df_B_correct.rename(columns={'index': 'Emotion Pair'}, inplace=True)

results_df_C_correct = pd.DataFrame({
    'Valence Test Statistic': valence_diff_df_C_correct['Test Statistic'],
    'Valence p-value': valence_diff_df_C_correct['p-value'],
    'Arousal Test Statistic': arousal_diff_df_C_correct['Test Statistic'],
    'Arousal p-value': arousal_diff_df_C_correct['p-value']
})
results_df_C_correct.reset_index(inplace=True)
results_df_C_correct.rename(columns={'index': 'Emotion Pair'}, inplace=True)

results_df_D_correct = pd.DataFrame({
    'Valence Test Statistic': valence_diff_df_D_correct['Test Statistic'],
    'Valence p-value': valence_diff_df_D_correct['p-value'],
    'Arousal Test Statistic': arousal_diff_df_D_correct['Test Statistic'],
    'Arousal p-value': arousal_diff_df_D_correct['p-value']
})
results_df_D_correct.reset_index(inplace=True)
results_df_D_correct.rename(columns={'index': 'Emotion Pair'}, inplace=True)

# Display the DataFrames to the user
dataframe=results_df_A_correct
dataframe=results_df_B_correct
dataframe=results_df_C_correct
dataframe=results_df_D_correct
results_df_A_correct, results_df_B_correct, results_df_C_correct, results_df_D_correct
```
