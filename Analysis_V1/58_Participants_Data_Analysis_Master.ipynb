```python
# Pie Chart
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_format
plt.pie([dep_tot, mild_dep_tot, non_dep_tot], colors=['r','b', 'y'], labels=['Depressed', 'Mild-Depressed','Non-Depressed'],
        autopct=autopct_format([dep_tot, mild_dep_tot, non_dep_tot]), pctdistance=0.85, explode=[0.05,0.05, 0.05]
        )

# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()

# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)

# Adding Title of chart
plt.title('Depressed vs Non-Depressed')
```
